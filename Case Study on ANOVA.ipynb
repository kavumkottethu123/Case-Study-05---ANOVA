### Check whether all the zones generate the same amount of sales. 

#### H0: Amount of sales for all the zones are same
#### H1: Amount of sales for all the zones  differs

```python
from scipy.stats import f_oneway
f,p=f_oneway(sales['Zone - A'],sales['Zone - B'],sales['Zone - C'],sales['Zone - D'])
```

```python
print("p-value for significance is: ", p)
print('statistic:',f)
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

#####  Since p-value is less than 0.05 , it rejects null hypothesis.So Amount of sales for all the zones differs or all the zones doesnot generates the same amount of sales
